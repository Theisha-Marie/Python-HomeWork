```python
#Step 8. Ops...it seems the index is from the most recent date. Make the first entry the oldest date.

import pandas as pd

# Load data from the provided URL
url = 'https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/09_Time_Series/Apple_Stock/appl_1980_2014.csv'
df = pd.read_csv(url)

# Convert 'Date' column to datetime type
df['Date'] = pd.to_datetime(df['Date'])

# Sort the DataFrame based on the 'Date' column
df.sort_values(by='Date', ascending=True, inplace=True)

# Reset the index to have a new order
df.reset_index(drop=True, inplace=True)

print(df)

```

```python
#Step 9. Get the last business day of each month
import pandas as pd



# Convert 'Date' column to datetime type and set it as the index
df['Date'] = pd.to_datetime(df['Date'])
df.set_index('Date', inplace=True)

# Get the last business day of each month
last_business_day_of_month = df.groupby(pd.Grouper(freq='BM')).last()


print(last_business_day_of_month)
```

```python
#Step 10. What is the difference in days between the first day and the oldest

import pandas as pd


# Convert 'Date' column to datetime type
df['Date'] = pd.to_datetime(df['Date'])

# Calculate the difference in days between the first day and the oldest date
difference_in_days = (df['Date'].max() - df['Date'].min()).days

print(difference_in_days)
```

```python
#Step 11. How many months in the data we have?

# Convert 'Date' column to datetime type
df['Date'] = pd.to_datetime(df['Date'])

# Calculate the number of unique months
num_months = df.resample('M', on='Date').size().shape[0]

print("Number of months in the data:", num_months)
```

```python
#Step 12. Plot the 'Adj Close' value. Set the size of the figure to 13.5 x 9 inches

import pandas as pd
import matplotlib.pyplot as plt

# Convert 'Date' column to datetime type
df['Date'] = pd.to_datetime(df['Date'])

# Set the 'Date' column as the index
df.set_index('Date', inplace=True)

# Plot 'Adj Close' values
plt.figure(figsize=(13.5, 9))
plt.plot(df.index, df['Adj Close'])
plt.xlabel('Date')
plt.ylabel('Adj Close')
plt.title('Apple Stock - Adj Close Value')
plt.grid(True)
plt.show()
```

```python
#BONUS: Create your own question and answer it.

import pandas as pd
import matplotlib.pyplot as plt

#30-day moving average of the 'Close' prices for Apple stock data

# Load data from the provided URL
url = 'https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/09_Time_Series/Apple_Stock/appl_1980_2014.csv'
df = pd.read_csv(url)


# Convert 'Date' column to datetime type
df['Date'] = pd.to_datetime(df['Date'])

# Set the 'Date' column as the index
df.set_index('Date', inplace=True)

# Calculate the 30-day moving average of 'Close' prices
df['30-day MA'] = df['Close'].rolling(window=30).mean()

# Plot 'Close' prices and the 30-day moving average
plt.figure(figsize=(13.5, 9))
plt.plot(df.index, df['Close'], label='Close')
plt.plot(df.index, df['30-day MA'], label='30-day MA', color='red')
plt.xlabel('Date')
plt.ylabel('Price')
plt.title('Apple Stock - Close Prices and 30-day Moving Average')
plt.legend()
plt.grid(True)
plt.show()

#The output will show a plot of the 'Close' prices for Apple stock along with the 30-day moving average.






```
