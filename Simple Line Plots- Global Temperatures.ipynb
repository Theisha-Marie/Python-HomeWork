```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file
csv_file_path = '/Users/theishamarie/Downloads/Global Temperature.csv'
data = pd.read_csv(csv_file_path)

# Check the column names
print(data.columns)  # Display the column names

# Assuming 'Monthly Anomaly' contains the temperature data
temperature_column_name = 'Monthly Anomaly'

# Extract years and temperature values from the dataset
years = data['Year']
temperatures = data[temperature_column_name]

# Create a figure and axis
plt.figure(figsize=(10, 6))
plt.title('Global Temperature Fluctuations Over Time')
plt.xlabel('Year')
plt.ylabel('Temperature Anomaly')

# Plot temperature fluctuations using different colors
plt.plot(years, temperatures, color='blue', label='Temperature Fluctuations')

# Customize the appearance of the plot
plt.grid(True)
plt.legend()
plt.tight_layout()

# Show the plot
plt.show()

```
